```python
# If the character name isn't in bold, use this function to get the name
# Print them out, with the episode number too to double check
def work_around(text, e):
    ret_text = ""
    for t in text[1:]:
        if t == ':':
            print(ret_text, e)
            return ret_text
        ret_text += t
        

# Use this function to separate the actions inside the lines
def separate_actions(text):
    is_action = False
    action = ""
    line = ""

    i = 0

    while i < len(text):
        if text[i] == '[':
            is_action = True

        elif text[i] == ']':
            is_action = False

        elif is_action:
            action += text[i]
        
        else:
            line += text[i]
        
        i += 1

    return " ".join(line.split()), " ".join(action.split())
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests as req

row_list = []

transcript_home_request = req.get('https://theinfosphere.org/Episode_Transcript_Listing')
transcript_home = BeautifulSoup(transcript_home_request.content, 'html.parser')

# Get the tv episodes, not the 4 movies at the end
episode_rows =  transcript_home.find_all('tr', class_='oCentre')[:-4]

base_url = "https://theinfosphere.org"
for e in episode_rows:

    # Get the information within the table cells
    e_cells = e.find_all('td')
    date = e_cells[1].text
    season = e_cells[3].text[1:3].lstrip('0')
    episode = e_cells[3].text[4:].lstrip('0')
    episode_num = e_cells[4].text

    # Get the link to the actual transcript
    transcript_request = req.get(base_url + e.find('td').find('a').get('href'))
    transcript = BeautifulSoup(transcript_request.content, 'html.parser')

    # Get each line, parse the character which is usually in bold text, and parse the line as well as any actions included in the line
    
    # Example:
    # Fry: Wow! [He turns to the man.] Um, can I have cuts?
    # Character: Fry, Action: He turns to the man., Line: Wow! Um, can I have cuts?
    for l in transcript.find_all('div', class_="poem"):
        character = l.find('b').text if l.find('b') else work_around(l.find('p').text, episode_num)
        for tag in l.find('p').find_all(['span', 'b', 'a']):
            tag.extract()
        line, action = separate_actions(l.text.replace('\n', '')[2:])

        row = {
            "date": date,
            "season": season,
            "episode": episode, 
            "character": character, 
            "line": line, 
            "action": action, 
            "episode_num": episode_num
        }
        row_list += [row]
    

df = pd.DataFrame(row_list)
df.to_csv('transcript_data.csv')
```
